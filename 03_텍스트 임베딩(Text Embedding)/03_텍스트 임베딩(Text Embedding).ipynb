# 정형 데이터와 비정형 데이터

## 정형 데이터 (structured data)
    테이블이나 관계형 데이터베이스 구조로 존재
	숫자나 텍스트, 날짜 등의 형식의 값을 가짐
	전체 데이터의 20% 가량을 차지
	저장 공간이 크지 않음
	분석이 쉽움

## 비정형 데이터 (unstructured data)
	정해진 구조가 없이 자유롭게 존재
	파일, 웹페이지, pdf 등의 형식으로 존재하고 텍스트, 이미지, 음성, 동영상 등등을 포함
	전체 데이터의 80% 가량을 차지
	상대적을 큰 저장 공간을 필요로 함
	분석이 어려움

## 데이터로 하고자 하는 것
    검색, 분석, 예측
    
    정형 데이터는 약간의 전처리만으로 원하는 결과를 얻기가 비교적 수월함
    비정형 데이터는 형식이 정해지지 않은 데이터이므로 하나로 통일하기 까다로움
    비정형 데이터를 정형화 시킨다면 그 과정에서 손실되는 데이터도 많이 생길 수 있으므로 비효율적

## Embedding의 사전적 의미

## 벡터 임베딩(Vector Embedding)
    비정형 데이터를 정보 손실을 최소화 하면서 하나의 형태로 만들기 위한 개념
    데이터를 벡터라는 숫자로 표현하는 것

### 벡터(Vector)
    크기와 방향이 있는 기하학적 개념
    벡터 공간 내에서 벡터간 사칙연산, 거리, 각도 계산이 가능

### 임베딩(Embedding)
    여러가지 형태의 데이터를 벡터 공간에 표현해서 벡터로 만드는 것(임베딩 벡터)
    텍스트 데이터를 임베딩 - 텍스트 임베딩
    이미지 데이터를 임베딩 - 이미지 임베딩

다루기 어려운 비정형 데이터를 벡터로 만들어 공간 내에서 가공하며 분석하는 것이 임베딩의 목적

## 좋은 임베딩
    데이터 분석에 도움이 되는 임베딩은 비슷한 성질의 데이터간 거리가 가까울 수록 좋음

### 단어 간 거리가 가까운 예시
	"왕"과 "여왕"은 의미적으로 매우 유사한 단어
	따라서 임베딩 공간에서 이 두 단어는 가까운 거리에 위치하게 되어야 함

### 단어 간 거리가 먼 예시
	"왕"과 "사과"는 의미적으로 전혀 관련이 없는 단어
	따라서 임베딩 공간에서 이 두 단어는 먼 거리에 위치하게 됨

이런 식으로 좋은 임베딩이 되면 가까운 벡터를 검색하거나 그룹화 하거나 특성 예측이 가능

## 임베딩 방법
	코사인유사도, 유클리드 거리 활용
	원핫 인코딩을 통한 원핫 벡터 이용
	딥러닝 기반

### 코사인 유사도 (Cosine Similarity)
    두 벡터 간의 각도를 이용하여 유사도를 측정하는 방법
    두 벡터가 이루는 각도의 코사인 값을 계산하여 유사도를 구함
    코사인 유사도는 -1에서 1 사이의 값을 가지며, 1에 가까울수록 두 벡터가 유사하다는 것을 의미

### 유클리드 거리 (Euclidean Distance)
    두 점 사이의 직선 거리를 계산하는 방법
    피타고라스 정리를 이용하여 두 점 사이의 거리를 구함
    유클리드 거리는 벡터 간의 거리를 측정하는 데 사용되며, 값이 작을수록 두 벡터가 유사하다는 것을 의미

### 원핫 인코딩 (One-Hot Encoding)
    각 단어를 고유한 인덱스로 변환하고, 해당 인덱스 위치에 1을 부여하고 나머지 위치에는 0을 부여하는 방식
    단어 집합이 [“사과”, “바나나”, “체리”]라면, 
    "사과"는 [1, 0, 0], "바나나"는 [0, 1, 0], "체리"는 [0, 0, 1]로 표현
    원핫 인코딩은 단어 간의 유사성을 반영하지 못함

### 딥러닝 기반 임베딩 (Deep Learning-based Embedding)
    인공신경망을 사용하여 단어를 벡터로 변환하는 방법
    대표적인 예로 Word2Vec, GloVe, BERT 등이 있음
    단어의 문맥적 의미를 반영하여 더 나은 성능을 보임
    딥러닝 기반 임베딩은 고차원 공간에서 단어 간의 유사성을 잘 반영할 수 있음

# 텍스트 임베딩(Text Embedding)
    텍스트 데이터를 고정된 크기의 벡터로 변환하여 수치화하는 작업
    텍스트 데이터를 머신러닝 모델의 입력으로 사용할 수 있음

## 다양한 텍스트 임베딩 방법

### TF-IDF(Term Frequency-Inverse Document Frequency): 
	단어의 빈도와 역 문서 빈도를 사용하여 각 단어의 중요도를 계산하는 방법
	주로 문서의 유사도를 구하거나, 검색 시스템에서 검색 결과의 중요도를 정하는 작업 등에 사용

### Word2Vec
	단어를 고정된 차원의 벡터로 변환하는 모델로, 단어 간의 문맥적 유사성을 반영
	CBOW와 Skip-gram 두 가지 방식이 있음

### GloVe (Global Vectors for Word Representation)
	전체 말뭉치의 통계 정보를 반영하여 단어 벡터를 생성
	단어 간의 전역적인 동시 발생 빈도를 이용

## FastText
	단어를 n-그램 단위로 분해하여 벡터화
	이를 통해 형태소 정보를 반영하고, OOV(Out-Of-Vocabulary) 문제를 해결

## ELMo (Embeddings from Language Models)
	문맥을 반영한 단어 임베딩을 생성
	양방향 LSTM을 사용하여 문맥 정보를 포함한 임베딩을 만듦

### BERT (Bidirectional Encoder Representations from Transformers)
	트랜스포머 기반 모델로, 양방향 문맥을 반영하여 단어의 의미를 이해
	다양한 NLP 작업에서 높은 성능을 보임

### GPT (Generative Pre-trained Transformer)
	언어 생성 모델로, 사전 학습된 트랜스포머를 사용하여 텍스트를 생성
	시퀀스의 다음 단어를 예측하는 방식으로 작동

### Doc2Vec
	문서 전체를 벡터로 변환하는 모델로, 문서 간의 유사성 비교 가능
	Word2Vec의 확장판

### USE (Universal Sentence Encoder)
	문장 또는 단락을 고정된 크기의 벡터로 인코딩
	다양한 텍스트 길이를 효율적으로 처리할 수 있음 있음수 

### 임베딩 예시

#### TF-IDF (Term Frequency-Inverse Document Frequency)
    TF-IDF는 텍스트의 단어 빈도(term frequency)와 역문서 빈도(inverse document frequency)를 계산하여 각 단어의 중요도를 측정하는 방법
    자주 등장하는 단어일수록 가중치가 높고, 여러 문서에 걸쳐 자주 등장하는 단어일수록 가중치가 낮아짐

```python
# 임베딩 예시
from sklearn.feature_extraction.text import TfidfVectorizer

# 예시 문서
documents = [
    "저는 오늘 아침에 커피를 마시며, 어제 친구와 함께 갔던 영화관에 대해 이야기했습니다.",
    "그 영화는 정말 재미있었고, 다시 보고 싶은 생각이 들었습니다."
]

# TF-IDF 벡터화
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)

# TF-IDF 행렬 출력
print("TF-IDF 행렬:")
print(tfidf_matrix.toarray())

# 단어 목록 출력
print("단어 목록:")
print(vectorizer.get_feature_names_out())
```

#### Word2Vec
    분포가설을 사용(어떤 단어의 의미는 단어가 같이 사용된 맥락에 의해 결정된다는 것)
    Word2Vec은 단어를 고정된 크기의 벡터로 변환하는 방법으로, 단어의 의미를 벡터 공간에서 표현할 수 있음
    주로 CBOW(Continuous Bag of Words)와 Skip-gram 모델을 사용
    
    CBOW
        주변 단어들(예: “오늘”, “아침”, “먹었다”)을 입력으로 받아 중심 단어(예: “토스트”)를 예측
        CBOW는 주변 단어들의 평균을 사용하여 중심 단어를 예측하는 모델
    
    Skip-Gram
        중심 단어를 입력으로 받아 주변 단어들을 예측
        대규모 데이터셋에서 더 좋은 성능을 보이는 경향이 있음
        
)


#### word2vec의 목표
    각 단어에 대해 고정된 크기의 벡터를 생성하는 것
    Word2Vec으로 만들어진 벡터는 단어의 의미적 특성을 반영하여 유사한 단어들은 벡터 공간에서 서로 가깝게 위치
    단어에 대한 사칙연산도 가능(단어의 임베딩이 벡터임으로 벡터의 사칙연산을 사용하는 것)

#### 버전 문제로 scipy 1.13버전에서 gensim에서 필요한 함수가 제거되어 1.12로 다운그레이드 필요


설치 후 주피터 커널 재시작이 필요할 수 있음

```python
from gensim.models import Word2Vec
from konlpy.tag import Okt

# 형태소 분석기 초기화
okt = Okt()

# 예시 문서
documents = [
    "저는 오늘 아침에 커피를 마시며, 어제 친구와 함께 갔던 영화관에 대해 이야기했습니다.",
    "그 영화는 정말 재미있었고, 다시 보고 싶은 생각이 들었습니다."
]

# 형태소 분석 및 토큰화
tokenized_documents = [okt.morphs(doc) for doc in documents]

# Word2Vec 모델 학습(sg=0: CBOW, sg=1: Skip-gram)
model = Word2Vec(sentences=tokenized_documents, vector_size=100, window=5, min_count=1, workers=4, sg=1)

# 단어 벡터 출력
word_vectors = model.wv
print("단어 '영화'의 벡터:")
print(word_vectors['영화'])

# 가장 유사한 단어 출력
print("단어 '영화'와 가장 유사한 단어들:")
print(word_vectors.similar_by_word('영화'))

# 모르는 단어와 유사한 단어 출력
try:
    print("단어 '녹차'와 가장 유사한 단어들:")
    print(word_vectors.similar_by_word('녹차'))
except KeyError as e:
    print(e)
```

#### 단점
    학습 데이터에 없는 단어(out-of-vocabulary)에 대해서는 임베딩을 구할 수 없음
    단어 임베딩은 단어별로 정의되기 때문에 다의어에 대한 처리가 어려움
    단어는 문맥에 따라 의미가 다를 수 있음
    
    예) 사과를 표현하는 apple과 회사를 표현하는 apple의 벡터 임베딩이 하나(두 단어를 구분할 수 없음)
    word2vec이 분포 가설을 제대로 반영하지 못하는 부분이라고 볼 수 있음


### FastText
    FastText는 Word2Vec의 확장
    단어 자체뿐만 아니라 단어를 구성하는 부분(서브워드, n-gram)을 벡터로 학습
    형태소가 유사한 단어들에 대한 더 나은 표현을 얻을 수 있음
    n-gram으로 분해하는 방식이므로 다의어도 문맥에 따라 적절한 의미 추론이 가능함

```python
from gensim.models import FastText
from konlpy.tag import Okt

# 형태소 분석기 초기화
okt = Okt()

# 예시 문서
documents = [
    "저는 오늘 아침에 커피를 마시며, 어제 친구와 함께 갔던 영화관에 대해 이야기했습니다.",
    "그 영화는 정말 재미있었고, 다시 보고 싶은 생각이 들었습니다."
]

# 형태소 분석 및 토큰화
tokenized_documents = [okt.morphs(doc) for doc in documents]

# FastText 모델 학습
model = FastText(sentences=tokenized_documents, vector_size=100, window=5, min_count=1, workers=4)

# 단어 벡터 출력
word_vectors = model.wv
print("단어 '영화'의 벡터:")
print(word_vectors['영화'])

# 가장 유사한 단어 출력
print("단어 '영화'와 가장 유사한 단어들:")
print(word_vectors.similar_by_word('영화'))

# 모르는 단어와 유사한 단어 출력
print("단어 '녹차'와 가장 유사한 단어들:")
print(word_vectors.similar_by_word('녹차'))
```

### BERT (Bidirectional Encoder Representations from Transformers)
    BERT는 사전 학습된 언어 모델로, 문맥을 고려하여 텍스트를 벡터로 변환할 수 있음
    단어뿐만 아니라 문장 단위의 벡터도 생성 가능

#### BertTokenizer.from_pretrained('bert-base-multilingual-cased')
    BERT 모델에 맞는 토크나이저를 로드
        BertTokenizer는 텍스트 데이터를 BERT 모델에 입력할 수 있는 형식으로 변환하는 역할
        텍스트를 토큰(단어 또는 서브워드 단위)으로 분할하고, 각 토큰을 BERT 모델에서 이해할 수 있는 정수 인덱스로 변환
        
    bert-base-multilingual-cased - 사전 훈련된 BERT 모델의 이름
        multilingual: 다국어를 지원하는 모델(영어 및 한국어, 중국어, 일본어, 프랑스어 등 100개 이상의 언어를 지원)
        cased: 대소문자를 구분하는 모델("Apple"과 "apple"을 다르게 인식)

#### TFBertModel.from_pretrained('bert-base-multilingual-cased')
    BERT 모델 자체를 로드(TFBertModel은 TensorFlow에서 BERT 모델을 구현한 클래스)
        from_pretrained('bert-base-multilingual-cased')를 통해 이미 학습된(pretrained) BERT 모델 가중치와 구조를 로드
    
    TFBertModel: BERT 모델의 기본 아키텍처를 제공
        로드된 모델은 입력된 텍스트를 기반으로 임베딩(벡터 표현)을 생성하고, 다양한 자연어 처리 작업에 사용할 수 있음

```python
#!pip install transformers
```

```python
import os
import warnings
from transformers import BertTokenizer, TFBertModel
import tensorflow as tf

os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'  # 대부분의 경고 메시지 억제

# 경고 메시지 무시
warnings.filterwarnings('ignore', category=FutureWarning)

# BERT 모델 및 토크나이저 로드
tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-cased')
model = TFBertModel.from_pretrained('bert-base-multilingual-cased')

# 예시 문서
text = "저는 오늘 아침에 커피를 마시며, 어제 친구와 함께 갔던 영화관에 대해 이야기했습니다."

# 토크나이징
inputs = tokenizer(text, return_tensors='tf')

# BERT 모델을 사용하여 임베딩 생성
outputs = model(**inputs)

# 문장 벡터 출력 (CLS 토큰 벡터 사용)
sentence_vector = outputs.last_hidden_state[:, 0, :].numpy()
print("문장 벡터:")
print(sentence_vector)
```

# inputs = tokenizer(text, return_tensors='tf')
    사전 학습된 BERT모델에 입력할 수 있는 형태로 변환하는 작업
    원본 텍스트를 토큰화하여 모델에 입력으로 사용할 수 있는 텐서로 변환한다.
    return_tensors='tf' 사용할 모델에 따라 'pt', 'np' 등 적용

## 변환 과정
    토큰화 (Tokenization)
        텍스트를 개별 토큰으로 나눔
        예) "저는 오늘 아침에 커피를 마시며"라는 문장은 ["저", "는", "오늘", "아침", "에", "커피", "를", "마시", "며"]로 나뉨

    토큰 ID 매핑 (Token IDs Mapping)
        사전 학습된 모델의 어휘(vocabulary)에 기반하여 각 토큰을 고유한 정수 ID로 변환

    특수 토큰 추가 (Special Tokens)
        BERT 모델은 문장의 시작에 [CLS] 토큰을, 끝에 [SEP] 토큰을 추가
    
    패딩 및 마스킹 (Padding and Masking)
        입력 시퀀스의 길이를 맞추기 위해 짧은 시퀀스는 패딩 토큰([PAD])을 추가하여 길이를 맞춤
        마스크 텐서를 생성하여 실제 토큰과 패딩 토큰을 구분

#### 반환 결과
    input_ids
        텍스트의 각 토큰을 고유 ID로 변환한 텐서
    attention_mask
        패딩 토큰([PAD])과 실제 토큰을 구분하는 마스크 텐서
        실제 토큰에는 1, 패딩 토큰에는 0이 할당
    token_type_ids
        문장의 구분을 위한 텐서
        단일 문장 입력의 경우, 모든 값이 0으로 설정

#### 토크나이징 예시

```python
from transformers import BertTokenizer

# BERT 토크나이저 로드
tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-cased')

# 예시 문서
text = "저는 오늘 아침에 커피를 마시며, 어제 친구와 함께 갔던 영화관에 대해 이야기했습니다."

# 토크나이징
inputs = tokenizer(text, return_tensors='tf')
inputs
```

```python
token_ids = inputs['input_ids'].numpy()[0]
token_ids
```

```python
# 토큰 및 ID 매핑 출력
tokens = tokenizer.convert_ids_to_tokens(token_ids)
token_id_pairs = list(zip(tokens, token_ids))

# 출력
for token, token_id in token_id_pairs:
    print(f"Token: {token}, ID: {token_id}")
```

### BERT 토크나이저에 의해 토큰화되고, 각 토큰이 고유 ID로 변환된 결과 input_ids
    각 ID는 BERT 모델의 어휘(vocabulary)에 있는 토큰에 매핑
    문장의 시작과 끝 등에 특수 토큰이 추가되고 원본 텍스트는 서브워드(subword) 단위로 토큰화 됨
    
#### BERT 모델의 주요 특수 토큰
    [CLS] 토큰 (ID: 101)
        문장의 시작을 나타내는 토큰
        주로 문장 수준의 작업(예: 문장 분류)에서 문장을 대표하는 임베딩 벡터를 생성하는 데 사용
    [SEP] 토큰 (ID: 102)
        문장의 끝을 나타내거나 문장 간의 구분을 나타내는 토큰
        BERT 모델은 두 개의 입력 문장을 구분하거나, 문장의 끝을 나타내기 위해 이 토큰을 사용
    [PAD] 토큰 (ID: 0)
        시퀀스의 길이를 맞추기 위해 패딩할 때 사용되는 토큰
        시퀀스의 실제 토큰과 패딩 토큰을 구분하기 위해 attention_mask와 함께 사용
    [MASK] 토큰 (ID: 103)
        마스킹 작업에서 사용되는 토큰
        BERT 모델의 사전 학습 단계에서 마스킹된 단어를 예측하는 작업(Masked Language Model, MLM)에서 사용
    

### outputs 객체
    outputs는 BERT 모델의 출력을 포함하는 객체
    
    일반적으로 다음과 같은 속성이 포함됨
        last_hidden_state
            BERT 모델의 마지막 은닉층(hidden layer)에서 생성된 임베딩 벡터
            (형태: (batch_size, sequence_length, hidden_size))
            각 토큰에 대한 벡터를 포함함
        pooler_output (옵션)
            [CLS] 토큰에 대한 임베딩 벡터
            문장 수준의 임베딩을 제공하며, 주로 문장 분류 작업에 사용
        hidden_states (옵션)
            BERT 모델의 모든 은닉층에서 생성된 임베딩 벡터
            각 층의 출력이 리스트 형태로 저장됨
            이 값을 얻으려면 임베딩 생성 시 output_hidden_states=True 설정 필요
                - outputs = model(**inputs, output_hidden_states=True)

#### 각 값은 다음과 같이 활용할 수 있음
    토큰 수준의 작업 - last_hidden_state
        예시: 개체명 인식(NER), 품사 태깅(POS tagging)
        각 토큰에 대한 임베딩 벡터가 필요
    
    문장 수준의 작업 - pooler_output
        예시: 문장 분류, 감정 분석
        문장 전체를 대표하는 임베딩 벡터가 필요
    
    고급 작업 - hidden_states
        예시: 층별 특성 추출, 임베딩 시각화, 모델 해석
        특정 층의 출력을 활용하여 모델의 동작을 더 깊이 이해하거나, 특정 층의 특성을 추출하기 위해

```python
# outputs의 구조 및 내용 확인
print(f"output의 type: {type(outputs)}")
print(f"output의 keys: {outputs.keys()}")
print(f"output.last_hidden_state.shape: {outputs.last_hidden_state.shape}")

if 'pooler_output' in outputs:
    print(f"outputs.pooler_output.shape: {outputs.pooler_output.shape}")
if 'hidden_states' in outputs:
    print(f"len(outputs.hidden_states): {len(outputs.hidden_states)}")
```

```python
# 문장 벡터 출력 (CLS 토큰 벡터 사용)
sentence_vector = outputs.last_hidden_state[:, 0, :].numpy() # outputs.last_hidden_state[0][0, :].numpy() 같은 결과
print("문장 벡터:")
print(sentence_vector)
```

#### outputs.last_hidden_state[:, 0, :].numpy() 해석
    [:, 0, :]
        텐서 슬라이싱 구문
        : - 전체 배치(배치 크기가 1이므로 하나의 문장만 처리)
        0 - 시퀀스의 첫 번째 토큰인 [CLS] 토큰을 선택(BERT 모델의 [CLS] 토큰은 문장 전체를 대표하는 임베딩 벡터이다.)
        : - 마지막 은닉 상태 벡터의 전체 차원을 의미

### 여러 문장에 대해 임베딩

```python
import os
from transformers import BertTokenizer, TFBertModel
import tensorflow as tf

# TensorFlow 경고 메시지 억제
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'

# BERT 모델 및 토크나이저 로드
tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-cased')
model = TFBertModel.from_pretrained('bert-base-multilingual-cased')

# 예시 문서들
texts = [
    "저는 오늘 아침에 커피를 마시며, 어제 친구와 함께 갔던 영화관에 대해 이야기했습니다.",
    "그 영화는 정말 재미있었고, 다시 보고 싶은 생각이 들었습니다.",
    "오늘 날씨가 너무 좋아서 산책을 다녀왔어요.",
    "BERT 모델은 자연어 처리에서 혁신적인 도구로 사용됩니다."
]

# 각 문장에 대해 임베딩 생성 및 출력
for i, text in enumerate(texts):
    # 토크나이징
    inputs = tokenizer(text, return_tensors='tf')

    # BERT 모델을 사용하여 임베딩 생성
    outputs = model(**inputs)

    # 문장 벡터 출력 (CLS 토큰 벡터 사용)
    sentence_vector = outputs.last_hidden_state[:, 0, :].numpy()
    
    print(f"문장 {i+1} 벡터:")
    print(sentence_vector)
    print("\n")  # 문장 구분을 위해 개행 추가
```

#### 결과
위와 같이 BERT 등의 임베딩 도구를 통해 임베딩이 완료된 벡터를 활용하여 원하는 작업에 사용할 수 있다.

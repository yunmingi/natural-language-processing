# 한국어의 특징
    한국어는 영어와는 다른 고유의 언어적 특성 때문에 자연어 처리(NLP)에서 많은 도전 과제를 안고 있다.
    한국어를 처리할 때 중요한 부분은 주로 형태소 분석, 조사 및 어미 처리, 그리고 한국어 특유의 문법 구조와 관련이 있다.

## 형태소 분석의 중요성
    형태소는 언어에서 의미를 가지는 가장 작은 단위이며 한국어에서는 특히 중요
    한국어 단어는 여러 형태소로 구성될 수 있으며, 이 형태소들이 결합하여 단어가 됨
    예) "가는길"
        "가" (동사) + "는" (현재형 어미) + "길" (명사)

### 형태소 분석이 중요한 이유
    1. 정확한 의미 파악
        형태소를 정확히 분석해야 문장의 정확한 의미를 파악할 수 있음
    2. 단어 생성
        한국어는 접사나 어미를 통해 새로운 단어를 쉽게 생성할 수 있어 형태소 분석을 통해 단어 분해 및 이해가 필요
    3. 다양한 형태의 단어 처리
        한국어는 다양한 어미 변형이 가능하므로 형태소 분석을 통해 이를 처리할 수 있어야 함       

#### 한국어 문장 변형 예시

접사 예시)
- 접사는 단어의 앞이나 뒤에 붙어서 의미를 추가하거나 변형하는 역할
  
    - 접두사  
        비: 비정상 (정상이 아닌)  
        불: 불행 (행복하지 않은)  
        미: 미완성 (완성되지 않은)  
    - 접미사  
        -적: 과학적 (과학과 관련된)  
        -스럽다: 사랑스럽다 (사랑이 느껴지는)  
        -기: 먹기 (먹다의 명사형)  

문장에 대한 다양한 어미 변형 예시)

- 어미는 동사나 형용사 같은 용언의 끝에 붙어 그 말의 문법적 기능을 나타내는 요소

     - 기본 문장: "공부하다"  
    
        - **평서문**  
            공부한다.  
            공부해요.  
            공부합니다.  
        - **의문문**  
            공부하니?  
            공부해요?  
            공부합니까?  
        - **명령문**  
            공부해라.  
            공부하세요.  
        - **청유문**  
            공부하자.  
            공부해요.  
        - **연결어미**  
            대등적 연결: 공부하고 (예: 나는 공부하고 운동을 한다.)  
            종속적 연결: 공부해서 (예: 열심히 공부해서 시험에 합격했다.)  
        - **전성어미**  
            명사형: 공부하기 (예: 공부하기가 쉽지 않다.)  
            부사형: 공부하게 (예: 열심히 공부하게 되었다.)  
            관형형: 공부하는 (예: 공부하는 학생들)
        - **높임말 어미**  
            존댓말: 공부하십니다, 공부하세요.  
            낮춤말: 공부해라, 공부해.  
        - **시제 어미**  
            현재시제: 공부한다, 공부해요.  
            과거시제: 공부했다, 공부했어요.  
            미래시제: 공부하겠다, 공부할 것이다.  

## 조사 및 어미 처리
    한국어에서는 조사와 어미가 매우 중요한 역할을 함
    조사는 주로 명사 뒤에 붙어 그 명사의 문법적 역할을 나타냄
    어미는 주로 동사나 형용사 뒤에 붙어 시제, 존대, 긍정/부정 등의 의미를 나타냄

    예시)
        조사: "책을 읽다"에서 "을"은 목적격 조사로, "책"이 목적어임을 나타냄
        어미: "읽었다"에서 "었다"는 과거 시제를 나타내는 어미

### 조사 및 어미 처리가 중요한 이유
    의미 차이
        조사의 변화는 문장의 의미를 크게 바꿀 수 있음
        "책이 좋다"와 "책을 좋다"는 의미가 다름

    문법적 정확성
        적절한 조사와 어미를 이해하고 처리해야 문법적으로 올바른 문장을 생성할 수 있음


### 한국어 자연어 처리의 도전 과제
    결국 한국어를 이용한 자연어 처리에서 가장 중요한 것은 형태소 분석과 조사 및 어미 처리에 있다.

    한국어가 교착어(agglutinative language)이기 때문에 가장 먼저 형태소 분석이 정확히 이루어져야 하며
    다음으로 어미나 조사가 어떻게 붙어있는지 의미를 잘 파악해야 이후의 작업이 진행될 수 있는 것이다.

    1. 형태소 분석
        문장의 정확한 의미 파악
    2. 다양한 조사의 처리
        다양한 조사와 그 변형을 정확히 이해하고 처리    3. 어미의 변화
        시제, 존대, 긍정/부정 등의 다양한 어미 변형을 이해하고 처리

## 한국어 자연어 처리에서 중요하게 생각해야 하는 부분
- **데이터의 전처리**

    - 데이터의 품질을 높이고, 분석의 정확성을 향상시킬 수 있음
- **한국어 특화된 도구 사용**  
    - 한국어에 특화된 형태소 분석기나 토크나이저를 사용하는 것이 중요
    - KoNLPy 라이브러리는 한국어 NLP를 위한 다양한 형태소 분석기(예: Mecab, Hannanum, Kkma, Komoran)를 제공
- **대용량 데이터 처리**  
    - 한국어 NLP 작업을 수행할 때, 대규모 데이터셋을 처리하는 능력 필요함
    - 대용량 데이터를 효율적으로 처리할 수 있는 알고리즘과 도구를 사용하는 것이 중요
- **문맥의 이해**  
    - 한국어는 문맥에 따라 의미가 달라질 수 있음
    - BERT와 같은 사전 학습된 언어 모델을 사용하면 문맥을 이해하고 처리하는 데 도움이 될 수 있음
- **특수한 한국어 처리 기술**  
    - 한국어 NLP에서는 고유명사 인식, 감정 분석, 문서 요약, 질의응답 등 다양한 작업이 필요함
    - 각각의 작업에 특화된 기술과 방법론이 필요할 수 있음

```python
# 예시 텍스트
text = """
저는 오늘 아침에 커피를 마시며, 어제 친구와 함께 갔던 영화관에 대해 이야기했습니다.
그 영화는 정말 재미있었고, 다시 보고 싶은 생각이 들었습니다.
"""
```

### 위 텍스트에서 한글 데이터 전처리 해보기

```python
from konlpy.tag import Okt
import re

# Okt 형태소 분석기 초기화
okt = Okt()
```

```python
# 텍스트 전처리
def preprocess_text(text):
    # 특수 문자 제거
    text = re.sub(r'[^\w\s]', '', text)
    # 숫자 변환 (여기서는 숫자를 제거)
    text = re.sub(r'\d+', 'NUM', text)
    # 불필요한 공백 제거
    text = re.sub(r'\s+', ' ', text).strip()
    return text

# 전처리된 텍스트
preprocessed_text = preprocess_text(text)
print("전처리된 텍스트:", preprocessed_text)
```

```python
# 형태소 분석
morphs = okt.morphs(preprocessed_text)
print("\n형태소:", morphs)
```

```python
# 품사 태깅
pos = okt.pos(preprocessed_text)
print("\n품사 태깅:", pos)
```

```python
# 명사 추출
nouns = okt.nouns(preprocessed_text)
print("\n명사:", nouns)
```

# 음성의 생성 과정

## 음성의 생성 메커니즘은 인간의 생리적 구조와 신경계의 협동 작용에 의해 이루어짐  

### 1. 호흡 메커니즘


    폐
        폐는 호흡을 통해 공기를 흡입하고 배출

    횡경막
        횡경막은 주된 호흡근으로, 수축하여 흉강을 확장시켜 공기를 들이마시고, 이완하여 흉강을 축소시켜 공기를 내뱉음
    
    늑간근
        늑간근 역시 흉강의 크기를 변화시키며 호흡을 도움

### 2. 발성 메커니즘


    성대
        성대는 후두에 위치한 두 개의 근육질 주름
        공기가 성대를 통과하면서 성대가 진동하여 소리가 발생

    성대 장력 조절
        성대 근육의 긴장과 이완을 통해 성대의 장력과 길이를 조절하여 음의 높낮이를 변화시킴

    글로탈 주기
        성대의 개폐 운동이 반복되면서 기본 주파수(기초 음)를 생성


### 3. 조음 메커니즘
    발음 기관
        소리가 입과 혀, 치아, 입술 등을 통해 변형됨
        
    주요 발음 기관
        - 혀
            혀의 위치와 모양은 다양한 모음과 자음을 형성
        - 입술
            입술의 개폐와 둥글림이 소리의 특성을 변화시킴
        - 치아
            치아와 혀, 입술의 상호작용은 특정 자음을 만듦
        - 연구개
            연구개의 위치는 비음을 조절

### 4. 공명 메커니즘
  

    
    공명강
        인두, 구강, 비강의 구조는 소리를 증폭하고 특유의 음색을 부여
        이 공명강의 형태와 크기에 따라 개인의 고유한 음색이 결정됨
    
    공명 효과
        공명이 잘 이루어지면 음성이 더 명확하고 풍부하게 들림

    신경계의 역할
        - 중추 신경계
            뇌는 음성 생성의 모든 과정을 조절
            말하기를 계획하고 조율하는 과정에서 대뇌 피질이 중요한 역할을 함
        
        - 말초 신경계
            뇌에서 내려온 신호는 말초 신경을 통해 발음 기관으로 전달
            이 신호는 근육의 움직임을 조절하여 음성을 생성

### 음성 생성 메커니즘 정리
    호흡: 횡경막과 늑간근의 협동으로 폐에서 공기를 들이마시고 내뱉음  
    발성: 공기가 성대를 통과하면서 성대가 진동하여 소리가 만들어짐  
    조음: 발음 기관이 소리를 변형하여 구체적인 음성을 만듦  
    공명: 공명강이 소리를 증폭하고 음색을 부여  
    신경 조절: 중추 및 말초 신경계가 모든 과정을 조절하여 조화로운 음성 생성을 가능하게 함

```python
import numpy as np
import matplotlib.pyplot as plt

# 아날로그 신호 시뮬레이션 (예: 440Hz 사인파)
sampling_rate = 44100  # 44.1kHz 샘플링 주파수
t = np.linspace(0, 0.01, int(0.01 * sampling_rate), endpoint=False)  # 10ms 동안의 신호
analog_signal = 0.5 * np.sin(2 * np.pi * 440 * t)  # 진폭 0.5, 주파수 440Hz의 사인파

# 새로운 샘플링 주파수 (예: 8000Hz)
new_sampling_rate = 8000  # 새로운 샘플링 주파수
num_samples = int(0.01 * new_sampling_rate)
sampled_t = np.linspace(0, 0.01, num_samples, endpoint=False)
sampled_signal = 0.5 * np.sin(2 * np.pi * 440 * sampled_t)  # 새로운 샘플링 주파수로 샘플링

# 양자화 함수 정의
def quantize(signal, bits):
    quantization_levels = 2 ** bits
    min_val = np.min(signal)
    max_val = np.max(signal)
    quantized_signal = np.round((signal - min_val) / (max_val - min_val) * (quantization_levels - 1))
    quantized_signal = quantized_signal / (quantization_levels - 1) * (max_val - min_val) + min_val
    return quantized_signal

# 4비트와 16비트로 양자화
quantized_signal_4bit = quantize(sampled_signal, 4)
quantized_signal_16bit = quantize(sampled_signal, 16)

# 원본 아날로그 신호와 샘플링된 신호, 양자화된 신호 시각화
plt.figure(figsize=(14, 10))

# 아날로그 신호
plt.subplot(4, 1, 1)
plt.plot(t, analog_signal, label='Analog Signal', alpha=0.75)
plt.title('Analog Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()

# 샘플링된 신호
plt.subplot(4, 1, 2)
plt.stem(sampled_t, sampled_signal, 'r', markerfmt='ro', basefmt=" ", label='Sampled Signal (8000 Hz)')
plt.title('Sampled Signal (8000 Hz)')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()

# 4비트 양자화된 신호
plt.subplot(4, 1, 3)
plt.stem(sampled_t, quantized_signal_4bit, 'g', markerfmt='go', basefmt=" ", label='Quantized Signal (4 bits)')
plt.plot(sampled_t, quantized_signal_4bit, 'g', alpha=0.5)  # 계단형 형태를 더 명확히 보기 위해 선 그래프 추가
plt.title('4-bit Quantized Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()

# 16비트 양자화된 신호
plt.subplot(4, 1, 4)
plt.stem(sampled_t, quantized_signal_16bit, 'b', markerfmt='bo', basefmt=" ", label='Quantized Signal (16 bits)')
plt.plot(sampled_t, quantized_signal_16bit, 'b', alpha=0.5)  # 계단형 형태를 더 명확히 보기 위해 선 그래프 추가
plt.title('16-bit Quantized Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 아날로그 신호 시뮬레이션 (예: 440Hz 사인파)
original_sampling_rate = 44100  # 원본 샘플링 주파수
t = np.linspace(0, 0.01, int(0.01 * original_sampling_rate), endpoint=False)  # 10ms 동안의 신호
analog_signal = 0.5 * np.sin(2 * np.pi * 440 * t)  # 진폭 0.5, 주파수 440Hz의 사인파

# 샘플링 주파수 목록
sampling_rates = [44100, 22050, 11025, 8000, 4000]

# 양자화 비트 수
quantization_bits_4 = 4
quantization_bits_16 = 16
quantization_levels_4 = 2 ** quantization_bits_4
quantization_levels_16 = 2 ** quantization_bits_16

plt.figure(figsize=(14, 18))

# 원본 아날로그 신호
plt.subplot(len(sampling_rates) + 1, 3, 1)
plt.plot(t, analog_signal, label='Analog Signal')
plt.title('Analog Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()

# 다양한 샘플링 주파수로 샘플링된 신호와 4비트 및 16비트 양자화된 신호
for i, sr in enumerate(sampling_rates, start=2):
    num_samples = int(len(t) * (sr / original_sampling_rate))
    sampled_t = np.linspace(0, 0.01, num_samples, endpoint=False)
    sampled_signal = 0.5 * np.sin(2 * np.pi * 440 * sampled_t)  # 샘플링

    # 4비트 양자화
    min_val = np.min(sampled_signal)
    max_val = np.max(sampled_signal)
    quantized_signal_4 = np.round((sampled_signal - min_val) / (max_val - min_val) * (quantization_levels_4 - 1))
    quantized_signal_4 = quantized_signal_4 / (quantization_levels_4 - 1) * (max_val - min_val) + min_val

    # 16비트 양자화
    quantized_signal_16 = np.round((sampled_signal - min_val) / (max_val - min_val) * (quantization_levels_16 - 1))
    quantized_signal_16 = quantized_signal_16 / (quantization_levels_16 - 1) * (max_val - min_val) + min_val

    # 샘플링된 신호 시각화
    plt.subplot(len(sampling_rates) + 1, 3, 3 * i - 2)
    plt.stem(sampled_t, sampled_signal, 'r', markerfmt='ro', basefmt=" ", label=f'Sampled Signal ({sr} Hz)')
    plt.title(f'Sampled Signal ({sr} Hz)')
    plt.xlabel('Time [s]')
    plt.ylabel('Amplitude')
    plt.grid()
    plt.legend()

    # 4비트 양자화된 신호 시각화
    plt.subplot(len(sampling_rates) + 1, 3, 3 * i - 1)
    plt.stem(sampled_t, quantized_signal_4, 'g', markerfmt='go', basefmt=" ", label=f'Quantized Signal (4 bits)')
    plt.title(f'Quantized Signal (4 bits)')
    plt.xlabel('Time [s]')
    plt.ylabel('Amplitude')
    plt.grid()
    plt.legend()

    # 16비트 양자화된 신호 시각화
    plt.subplot(len(sampling_rates) + 1, 3, 3 * i)
    plt.stem(sampled_t, quantized_signal_16, 'b', markerfmt='bo', basefmt=" ", label=f'Quantized Signal (16 bits)')
    plt.title(f'Quantized Signal (16 bits)')
    plt.xlabel('Time [s]')
    plt.ylabel('Amplitude')
    plt.grid()
    plt.legend()

plt.tight_layout()
plt.show()
```

## 오디오 파일의 디지털 표현
    샘플링
      오디오 신호는 일정한 시간 간격으로 샘플링
      샘플링 주파수가 높을수록 원본 아날로그 신호를 더 정확하게 재현가능
      
    양자화
      샘플링된 각 신호 값은 이산적인 디지털 값으로 변환됨
      양자화 비트 수가 많을수록 신호를 더 세밀하게 표현가능

    디지털 신호
      결과적으로 오디오 파일은 샘플링된 시간 지점마다 양자화된 진폭 값을 저장한 데이터로 구성

### 오디오 파일로부터 샘플링 및 양자화 예시

```python
import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf

# 오디오 파일 로드
file_path = '녹음.wav'
signal, sampling_rate = sf.read(file_path)

# 구간 선택 (예: 처음 2sec)
duration = 2.0  # 10ms
num_samples = int(duration * sampling_rate)
t = np.linspace(0, duration, num_samples, endpoint=False)
signal_segment = signal[:num_samples]

# 양자화 함수 정의
def quantize(signal, bits):
    quantization_levels = 2 ** bits
    min_val = np.min(signal)
    max_val = np.max(signal)
    quantized_signal = np.round((signal - min_val) / (max_val - min_val) * (quantization_levels - 1))
    quantized_signal = quantized_signal / (quantization_levels - 1) * (max_val - min_val) + min_val
    return quantized_signal

# 4비트와 16비트로 양자화
quantized_signal_4bit = quantize(signal_segment, 4)
quantized_signal_16bit = quantize(signal_segment, 16)

# 원본 신호와 양자화된 신호 시각화
plt.figure(figsize=(14, 10))

# 원본 신호
plt.subplot(3, 1, 1)
plt.plot(t, signal_segment, label='Original Signal', alpha=0.75)
plt.title('Original Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()

# 4비트 양자화된 신호
plt.subplot(3, 1, 2)
plt.stem(t, quantized_signal_4bit, 'g', markerfmt='go', basefmt=" ", label='Quantized Signal (4 bits)')
plt.title('4-bit Quantized Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()

# 16비트 양자화된 신호
plt.subplot(3, 1, 3)
plt.stem(t, quantized_signal_16bit, 'b', markerfmt='bo', basefmt=" ", label='Quantized Signal (16 bits)')
plt.title('16-bit Quantized Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```

## 딥러닝 모델에서의 데이터 처리
    딥러닝 모델은 이러한 디지털 신호 데이터를 입력으로 받아 학습

- 딥러닝 모델이 학습하는 일반적인 단계
1. 전처리: 원본 오디오 데이터를 전처리하여 딥러닝 모델에 적합한 형태로 변환
2. 노이즈 제거: 오디오 신호에서 잡음을 제거
3. 정규화: 신호의 진폭 값을 일정한 범위로 정규화
4. 특징 추출: 오디오 신호에서 주파수 도메인 특성(MFCC, 스펙트로그램 등)을 추출
5. 모델 학습: 전처리된 데이터를 딥러닝 모델에 입력하여 학습
6. 특징 학습: 모델이 오디오 데이터의 중요한 특징을 학습
7. 패턴 인식: 모델이 학습된 특징을 바탕으로 패턴을 인식
